```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from scipy.special import expit  # For sigmoid activation

# Load data
y_true = np.load('/content/drive/My Drive/PROJECT/y_val1.npy')
y_scores = np.load('/content/drive/My Drive/PROJECT/trained_aunet_model_predictions-2.npy')

# Convert `y_true` to binary (0 or 1)
y_true = (y_true >= 0.5).astype(int)

# Ensure `y_scores` are valid probabilities (0 to 1)
if y_scores.min() < 0 or y_scores.max() > 1:
    y_scores = expit(y_scores)  # Apply sigmoid

# Flatten arrays to avoid shape mismatch
y_true = y_true.flatten()
y_scores = y_scores.flatten()

# Compute ROC curve and AUC
fpr, tpr, _ = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Plot AUC-ROC Curve
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUC-ROC Curve')
plt.legend()
plt.show()
```
